1. Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:

   - $f(x) = ln(x), x_0 = 1$

      $f´(x) = \frac{1}{x}$

      $f´´(x) = \frac{-1}{x^2}$

      $f´´´(x) = \frac{2}{x^3}$

      $f(x) \approx f(x_0) + f´(x)(x - x_0) + \frac{f´´(x_0)}{2!} (x - x_0)^2 + \frac{f´´´(x_0)}{3!} (x - x_0)^3$

      $\approx ln(1) + (x -1) - \frac{(x - 1)^2}{2} + \frac{2}{6}(x - 1)^3$

      $\approx \frac{1}{3}(x - 1)^3 - \frac{(x - 1)^2}{2} + (x - 1)$
      

      


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 12, 800)

x_pos = np.linspace(0.1, 12, 800)

f = np.log(x_pos)
g = x - 1
h = -0.5*(x - 1)**2 + (x - 1)
q = (1/3)*(x - 1)**3 - 0.5*(x - 1)**2 + (x - 1)

plt.figure(figsize=(12, 7))

plt.plot(x_pos, f, color='green', linewidth=2, label=r'$f(x)=\ln(x)$')
plt.plot(x, g, color='red', linewidth=2, label=r'$g(x)=x-1$')
plt.plot(x, h, color='blue', linewidth=2, label=r'$h(x)=-\frac{(x-1)^2}{2}+x-1$')
plt.plot(x, q, color='purple', linewidth=2, label=r'$q(x)=\frac{1}{3}(x-1)^3 -\frac{(x-1)^2}{2}+x-1$')

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)

plt.xlim(-12, 12)
plt.ylim(-8, 8)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Funciones y aproximaciones alrededor de $x=1$')
plt.legend()

plt.show()
```

- $x_0 = 10$

    $f(x) = \ln(x) $

    $f'(x) = \frac{1}{x}$

    $f''(x) = -\frac{1}{x^2}$

    $f'''(x) = \frac{2}{x^3}$

    $f(x)\approx f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2!}(x-x_0)^2+\frac{f'''(x_0)}{3!}(x-x_0)^3$

    $\approx \ln(10)+\frac{1}{10}(x-10) -\frac{1}{200}(x-10)^2 +\frac{1}{3000}(x-10)^3$

    $\approx \frac{1}{3000}(x-10)^3 -\frac{1}{200}(x-10)^2 +\frac{1}{10}(x-10) + \ln(10)$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 20, 800)

x_pos = np.linspace(1, 20, 800)

f = np.log(x_pos)

T1 = np.log(10) + (1/10)*(x - 10)
T2 = np.log(10) + (1/10)*(x - 10) - (1/200)*(x - 10)**2
T3 = np.log(10) + (1/10)*(x - 10) - (1/200)*(x - 10)**2 + (1/3000)*(x - 10)**3

plt.figure(figsize=(12, 7))

plt.plot(x_pos, f, color='green', linewidth=2,
         label=r'$f(x)=\ln(x)$')

plt.plot(x, T1, color='red', linewidth=2,
         label=r'$T_1(x)=\ln(10)+\frac{1}{10}(x-10)$')

plt.plot(x, T2, color='blue', linewidth=2,
         label=r'$T_2(x)=\ln(10)+\frac{1}{10}(x-10)-\frac{1}{200}(x-10)^2$')

plt.plot(x, T3, color='purple', linewidth=2,
         label=r'$T_3(x)=\ln(10)+\frac{1}{10}(x-10)-\frac{1}{200}(x-10)^2+\frac{1}{3000}(x-10)^3$')

# Ejes
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)

plt.xlim(1, 20)
plt.ylim(-1, 4)

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.title(r'Polinomios de Taylor de $\ln(x)$ alrededor de $x_0=10$')

plt.legend(fontsize=10)

plt.show()
```

      $f(x) \approx f(0) + f'(0)(x - 0) + \frac{f''(0)}{2}(x - 0)^2 + \frac{f'''(0)}{6}(x - 0)^3$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 800)

f = np.cos(x)

T0 = np.ones_like(x)                 
T1 = np.ones_like(x)                
T2 = 1 - (x**2)/2                    
T3 = 1 - (x**2)/2                   

plt.figure(figsize=(12, 7))

plt.plot(x, f, color='green', linewidth=2,
         label=r'$f(x)=\cos(x)$')

plt.plot(x, T1, color='red', linewidth=2,
         label=r'$T_1(x)=1$')

plt.plot(x, T3, color='purple', linewidth=2,
         label=r'$T_3(x)=1-\frac{x^2}{2}$')

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)

plt.xlim(-5, 5)
plt.ylim(-2, 2)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'Polinomios de Taylor de $\cos(x)$ alrededor de $x_0=0$')

plt.legend(fontsize=10)

plt.show()
```

- $\frac{1}{(1 - x)}, x_0 = 0.5$

    $f(x) = (1-x)^{-1}$

    $f'(x) = (1-x)^{-2}$

    $f''(x) = 2(1-x)^{-3}$

    $f'''(x) = 6(1-x)^{-4}$

    $f(x) \approx f(0.5) + f'(0.5)(x - 0.5) + \frac{f''(0.5)}{2} (x - 0.5)^2 + \frac{f'''(0.5)}{6} (x - 0.5)^3$

    $f(x) \approx 2 + 4(x-0.5) + \frac{16}{2}(x-0.5)^2 + \frac{96}{6}(x-0.5)^3$

    $f(x) \approx 16(x-0.5)^3 + 8(x-0.5)^2 + 4(x-0.5) + 2$

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 0.5

x = np.linspace(-2, 0.9, 800)

f = 1 / (1 - x)

T1 = 2 + 4*(x - x0)
T2 = 2 + 4*(x - x0) + 8*(x - x0)**2
T3 = 2 + 4*(x - x0) + 8*(x - x0)**2 + 16*(x - x0)**3

plt.figure(figsize=(12, 7))

plt.plot(x, f, color='green', linewidth=2,
         label=r'$f(x)=\frac{1}{1-x}$')

plt.plot(x, T1, color='red', linewidth=2,
         label=r'$T_1(x)=2+4(x-0.5)$')

plt.plot(x, T2, color='blue', linewidth=2,
         label=r'$T_2(x)=2+4(x-0.5)+8(x-0.5)^2$')

plt.plot(x, T3, color='purple', linewidth=2,
         label=r'$T_3(x)=2+4(x-0.5)+8(x-0.5)^2+16(x-0.5)^3$')

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)

plt.xlim(-2, 0.9)
plt.ylim(-10, 40)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'Polinomios de Taylor de $\frac{1}{1-x}$ alrededor de $x_0=0.5$')

plt.legend(fontsize=10)

plt.show()
```
